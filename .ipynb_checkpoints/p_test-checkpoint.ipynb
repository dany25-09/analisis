### 3. Intervalo de confianza
El intervalo de confianza (IC) es esencial en las estad칤sticas y muy importante para los cient칤ficos de datos. 
Como suena, el intervalo de confianza es un rango de valores. En condiciones ideales, debe contener la mejor estimaci칩n de un par치metro estad칤stico. Se expresa como porcentaje. El intervalo de confianza del 95% es el m치s com칰n. Puede utilizar otros valores como 97%, 90%, 75% o incluso 99% de intervalo de confianza si su investigaci칩n lo exige.

En las siguientes l칤neas vamos a calcular el intervalo de confianza para el valor de flete medio de todos los env칤os

Primero obtener los valores necesarios para su c치lculo:
* 洧녦=洧洧뉧롐놿롐뒳롐 洧녬洧 洧녳洧녩 洧녴洧녹洧뉧롐멇롐뫯롐洧녩 
* 洧띺=洧녡洧녰洧녮洧녵洧녰洧녭洧녰洧녫洧녩洧녵洧녫洧녰洧녩
* 洧랥=洧냥洧뉧롐멇롐洧녰洧녩洧녫洧녰칩洧녵 洧냦洧멇롐뫯롐뀛롐洧녬洧녩洧륲n* 洧녵=洧녢洧녩洧녴洧녩침洧녶 洧녬洧 洧녳洧녩 洧녴洧녹洧뉧롐멇롐뫯롐洧녩 
* 洧녾=洧녤洧녩洧녳洧녶洧 洧녬洧 洧녾洧녢洧뉧롐멇롐

```python
#Media de la muestra
x=data['valor_flete'].mean()
#significancia
alpha=0.05 #para tener el 95% de confianza
# Desviaci칩n est치ndar
sigma=data['valor_flete'].std()
# Tama침o de la muestra
n=len(data['valor_flete'])
#Valor del test 
z=1.96
```

El valor de z con ese nivel de significancia se puede obtener de la siguiente tabla, este depende exclusivamente del nivel de  significacncia que se elija

[![0*i0yTYIZw0pgaZAsh.png](https://cdn-images-1.medium.com/max/800/0*i0yTYIZw0pgaZAsh.png)]()

Con los valores de construcci칩n del intervalo es posible calcular f치cilmente los intervalos con la siguiente f칩rmula

```python
#error est치ndar
error_estandar = sigma /np.sqrt(n)
#c치lculo intervalos de confianza
limite_inferior = x - z* error_estandar  
limite_superior = x + z* error_estandar
#Almacenar los dos l칤mites del intervalo
intervalo=[limite_inferior, limite_superior]
intervalo
```

Esto nos indica que el valor real del promedio poblacional(que es desconocido porque trabajamos conuna muestra) est치 entre 20.3 y 21.2 con un 95% de confianza

### 4. Validaci칩n de hip칩tesis
La prueba de hip칩tesis es un m칠todo estad칤stico que se utiliza para tomar decisiones estad칤sticas utilizando datos experimentales. La prueba de hip칩tesis es b치sicamente una suposici칩n que hacemos sobre el par치metro de poblaci칩n.

Una prueba de hip칩tesis eval칰a dos enunciados mutuamente excluyentes sobre una poblaci칩n para determinar qu칠 enunciado est치 mejor respaldado por los datos de muestra. Cuando decimos que un hallazgo es estad칤sticamente significativo, es gracias a una prueba de hip칩tesis.

### 4.1. Hip칩tesis

**Hip칩tesis nula:** Es una declaraci칩n general o posici칩n predeterminada de que no existe relaci칩n entre dos fen칩menos medidos, o no hay asociaci칩n entre grupos.
En otras palabras, es una suposici칩n b치sica o se basa en el conocimiento del dominio o del problema.

**Hip칩tesis alternativa:** Es la hip칩tesis utilizada en la prueba de hip칩tesis que es contraria a la hip칩tesis nula. Por lo general, se considera que las observaciones son el resultado de un efecto real (con cierta cantidad de variaci칩n aleatoria superpuesta)

### 5. Z-test
Una prueba z es una prueba estad칤stica que se utiliza para determinar si dos medias poblacionales son diferentes cuando se conocen las varianzas y el tama침o de la muestra es grande. Se supone que la estad칤stica de la prueba tiene una distribuci칩n normal y se deben conocer los par치metros, como la desviaci칩n est치ndar, para poder realizar una prueba z precisa.

Supuestos importantes:

* Sus datos deben seleccionarse al azar de una poblaci칩n, donde cada elemento tiene la misma probabilidad de ser seleccionado.
* Los tama침os de las muestras deben ser iguales si es posible.


En este caso vamos a tomar dos regiones de env칤o para revisar la media muestral a trav칠s de este test  

```python
# Elegir las ciudades
data_flete=data.groupby('departamento_vendedor').mean().reset_index()[['departamento_vendedor','valor_flete']]
#graficar lasciudades por promedios
data_flete.plot.bar(x='departamento_vendedor', y='valor_flete')
```

Por la cantidad de datos de los departamento de PR y RJ adem치s por la cercan칤a de los promedios se deciden elegir estas dos zonas para revisi칩n si el par치metro poblacional del promedio del valor del flete es igual (ya que en este momento solamente contamos con el par치metro muestral). As칤 nuestra prueba de hip칩tesis ser치:

* H0 : El promedio de los dos grupos son iguales
* H1 : El promedio de los dos grupos **NO son iguales**


```python
#Mostrar promedio de GO y MA
flete_PR=data[data['departamento_vendedor']=='PR']['valor_flete']
flete_RJ=data[data['departamento_vendedor']=='RJ']['valor_flete']
print('Promedio GO: '+str(flete_PR.mean()))
print('Promedio MA: '+str(flete_RJ.mean()))
```

```python
# Valores de z y p
ztest ,pval1 = stests.ztest(flete_PR, x2=flete_RJ, value=0,alternative='two-sided')
print('Valor de z: '+str(ztest))
print('Valor p: '+ str(pval1))
```

```python
# Regla de decisi칩n si el p valor es menor que la significancia escogida 
confianza=0.95
significancia=1-confianza
# Si se tiene menor valor se rechaza la hip칩tesis nula
if pval1<significancia:
    print("Rechazar hip칩tesis nula")
else:
    print("No rechazar hip칩tesis nula")
```

Por lo tanto podemos inferir con un nivel de confianza del 95% que los promedios del valor de flete de estas dos zonas no es igual
